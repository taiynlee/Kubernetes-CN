```
# Series(1D).replace(轉換字典)
# Series: unique, value_counts
train_df["作者"].value_counts()
```

```
u=train_df["作者"].unique()
print(u)
```

```
trans={n:i for i,n in enumerate(u)}
trans_r={i:n for i,n in enumerate(u)}
y_train=train_df["作者"].replace(trans)
y_test=test_df["作者"].replace(trans)
y_test
```

```
u
trans
```

```
import jieba
p=train_df["內容"][0]
" ".join(jieba.cut(p))
```

```
# 看情境 要不要去換行 要不要去標點符號 要不要去除常用詞(基本上不用 因為演算法有處理)
# p.replace("\r","").replace("\n","")
```

```
# apply(func)
def poemcut(p):
  p=" ".join(jieba.cut(p))
  return p.replace("\r","").replace("\n","")
x_train=train_df["內容"].apply(poemcut)
x_test=test_df["內容"].apply(poemcut)
x_test
```

```
# fit 找出最大值與最小值 不像其他的訓練 transfeorm才是真正轉為數字
# 訓練資料 fit+transform 測試資料只做transform 測試資料遇到沒看過的資料 直接不理
from sklearn.feature_extraction.text import CountVectorizer
vec=CountVectorizer()
x_train_vec=vec.fit_transform(x_train)
x_test_vec=vec.transform(x_test)
x_train_vec
# sparse matrix 只儲存非0的資料
```

```
print(x_train_vec)
```

```
vec.vocabulary_
```

```
# vec.vocabulary_["。"]
# 檢查標點有沒有去掉
```

```
from sklearn.naive_bayes import MultinomialNB
clf=MultinomialNB()
clf.fit(x_train_vec,y_train)
```

```
pre=clf.predict(x_test_vec)
```

```
from sklearn.metrics import accuracy_score
accuracy_score(pre,y_test)
```

```
from sklearn.metrics import confusion_matrix
mat=confusion_matrix(y_test,pre)
pd.DataFrame(mat,
             columns=[n+"(預測值)" for n in u],
             index=[n+"(真實值)" for n in u])
```
